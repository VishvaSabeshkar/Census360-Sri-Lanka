```sql
 -- Step 1: Verify dashboard source tables

SELECT 
  'gold_dashboard_kpi_summary' AS table_name,
  COUNT(*) AS row_count
FROM workspace.census360.gold_dashboard_kpi_summary

UNION ALL

SELECT 
  'gold_district_telco_summary' AS table_name,
  COUNT(*) AS row_count
FROM workspace.census360.gold_district_telco_summary

UNION ALL

SELECT 
  'gold_ds_telco_summary' AS table_name,
  COUNT(*) AS row_count
FROM workspace.census360.gold_ds_telco_summary

UNION ALL

SELECT 
  'gold_telco_opportunity_ranking' AS table_name,
  COUNT(*) AS row_count
FROM workspace.census360.gold_telco_opportunity_ranking

UNION ALL

SELECT 
  'gold_gn_telco_scored_v2' AS table_name,
  COUNT(*) AS row_count
FROM workspace.census360.gold_gn_telco_scored_v2

UNION ALL

SELECT 
  'gold_ml_opportunity_predictions' AS table_name,
  COUNT(*) AS row_count
FROM workspace.census360.gold_ml_opportunity_predictions

UNION ALL

SELECT 
  'gold_ml_model_comparison' AS table_name,
  COUNT(*) AS row_count
FROM workspace.census360.gold_ml_model_comparison;
```

```sql
-- Step 2: National KPI Summary for dashboard cards

SELECT
  total_provinces,
  total_districts,
  total_ds_divisions,
  total_gn_divisions,
  total_population,
  total_housing_units,
  ROUND(avg_household_size, 2) AS avg_household_size,
  ROUND(avg_telco_demand_opportunity_score, 2) AS avg_telco_demand_opportunity_score,
  very_high_opportunity_gns,
  high_opportunity_gns,
  medium_opportunity_gns,
  low_opportunity_gns
FROM workspace.census360.gold_dashboard_kpi_summary;
```

```sql
-- Step 3: Opportunity Tier Distribution

SELECT
  opportunity_tier,
  COUNT(*) AS gn_division_count,
  ROUND(
    COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (),
    2
  ) AS percentage_of_gns
FROM workspace.census360.gold_gn_telco_scored_v2
GROUP BY opportunity_tier
ORDER BY 
  CASE opportunity_tier
    WHEN 'Very High' THEN 1
    WHEN 'High' THEN 2
    WHEN 'Medium' THEN 3
    WHEN 'Low' THEN 4
    ELSE 5
  END;
```

```sql
-- Step 4: District Opportunity Ranking

SELECT
  province_name,
  district_name,
  gn_division_count,
  district_population_total,
  district_housing_units,
  ROUND(avg_household_size, 2) AS avg_household_size,
  ROUND(avg_mobile_data_demand_score, 2) AS avg_mobile_data_demand_score,
  ROUND(avg_home_broadband_opportunity_score, 2) AS avg_home_broadband_opportunity_score,
  ROUND(avg_youth_package_opportunity_score, 2) AS avg_youth_package_opportunity_score,
  ROUND(avg_rural_coverage_priority_score, 2) AS avg_rural_coverage_priority_score,
  ROUND(avg_telco_demand_opportunity_score, 2) AS avg_telco_demand_opportunity_score,
  ROUND(max_gn_opportunity_score, 2) AS max_gn_opportunity_score,
  district_opportunity_tier
FROM workspace.census360.gold_district_telco_summary
ORDER BY avg_telco_demand_opportunity_score DESC;
```

```sql
-- Step 5B: Top 20 DS Divisions by Telco Opportunity

SELECT
  province_name,
  district_name,
  ds_division_name,
  gn_division_count,
  ds_population_total,
  ds_housing_units,
  ROUND(avg_telco_demand_opportunity_score, 2) AS avg_telco_demand_opportunity_score,
  ROUND(max_gn_opportunity_score, 2) AS max_gn_opportunity_score,
  ds_opportunity_tier
FROM workspace.census360.gold_ds_telco_summary
ORDER BY avg_telco_demand_opportunity_score DESC
LIMIT 20;
```

```sql
-- Step 6: GN-Level Opportunity Ranking

SELECT
  national_opportunity_rank,
  province_name,
  district_name,
  ds_division_name,
  gn_division_name,
  gnd_uid,
  population_total,
  occupied_housing_units,
  ROUND(household_size, 2) AS household_size,
  ROUND(mobile_data_demand_score, 2) AS mobile_data_demand_score,
  ROUND(home_broadband_opportunity_score, 2) AS home_broadband_opportunity_score,
  ROUND(youth_package_opportunity_score, 2) AS youth_package_opportunity_score,
  ROUND(rural_coverage_priority_score, 2) AS rural_coverage_priority_score,
  ROUND(telco_demand_opportunity_score, 2) AS telco_demand_opportunity_score,
  opportunity_tier
FROM workspace.census360.gold_telco_opportunity_ranking
ORDER BY national_opportunity_rank ASC;
```

```sql
-- Step 7: Top 20 GN Divisions by Mobile Data Demand Opportunity

SELECT
  province_name,
  district_name,
  ds_division_name,
  gn_division_name,
  population_total,
  ROUND(working_age_share, 4) AS working_age_share,
  ROUND(child_share, 4) AS child_share,
  ROUND(mobile_data_demand_score, 2) AS mobile_data_demand_score,
  opportunity_tier
FROM workspace.census360.gold_gn_telco_scored_v2
ORDER BY mobile_data_demand_score DESC
LIMIT 20;
```

```sql
-- Step 8: Top 20 GN Divisions by Home Broadband Opportunity

SELECT
  province_name,
  district_name,
  ds_division_name,
  gn_division_name,
  occupied_housing_units,
  ROUND(household_size, 2) AS household_size,
  population_total,
  ROUND(home_broadband_opportunity_score, 2) AS home_broadband_opportunity_score,
  opportunity_tier
FROM workspace.census360.gold_gn_telco_scored_v2
ORDER BY home_broadband_opportunity_score DESC
LIMIT 20;
```

```sql
-- Step 9: Top 20 GN Divisions by Youth Package Opportunity

SELECT
  province_name,
  district_name,
  ds_division_name,
  gn_division_name,
  population_total,
  age_0_14,
  age_15_59,
  ROUND(child_share, 4) AS child_share,
  ROUND(working_age_share, 4) AS working_age_share,
  ROUND(youth_package_opportunity_score, 2) AS youth_package_opportunity_score,
  opportunity_tier
FROM workspace.census360.gold_gn_telco_scored_v2
ORDER BY youth_package_opportunity_score DESC
LIMIT 20;
```

```sql
-- Step 10: Top 20 GN Divisions by Rural Coverage Priority

SELECT
  province_name,
  district_name,
  ds_division_name,
  gn_division_name,
  population_total,
  occupied_housing_units,
  ROUND(household_size, 2) AS household_size,
  ROUND(rural_coverage_priority_score, 2) AS rural_coverage_priority_score,
  opportunity_tier
FROM workspace.census360.gold_gn_telco_scored_v2
ORDER BY rural_coverage_priority_score DESC
LIMIT 20;
```

```sql
-- Step 11: Top 20 GN Divisions by ML Predicted Opportunity Score

SELECT
  province_name,
  district_name,
  ds_division_name,
  gn_division_name,
  population_total,
  occupied_housing_units,
  ROUND(actual_opportunity_score, 2) AS actual_opportunity_score,
  ROUND(predicted_opportunity_score, 2) AS predicted_opportunity_score,
  ROUND(prediction_error, 4) AS prediction_error,
  opportunity_tier,
  ml_prediction_tier
FROM workspace.census360.gold_ml_opportunity_predictions
ORDER BY predicted_opportunity_score DESC
LIMIT 20;
```

```sql
-- Step 12: ML Actual vs Predicted Opportunity Score

SELECT
  province_name,
  district_name,
  ds_division_name,
  gn_division_name,
  ROUND(actual_opportunity_score, 2) AS actual_opportunity_score,
  ROUND(predicted_opportunity_score, 2) AS predicted_opportunity_score,
  ROUND(prediction_error, 4) AS prediction_error,
  opportunity_tier,
  ml_prediction_tier
FROM workspace.census360.gold_ml_opportunity_predictions
ORDER BY actual_opportunity_score DESC;
```

```sql
-- Step 12B: Top 20 ML Prediction Comparison

SELECT
  province_name,
  district_name,
  ds_division_name,
  gn_division_name,
  ROUND(actual_opportunity_score, 2) AS actual_opportunity_score,
  ROUND(predicted_opportunity_score, 2) AS predicted_opportunity_score,
  ROUND(prediction_error, 4) AS prediction_error,
  opportunity_tier,
  ml_prediction_tier
FROM workspace.census360.gold_ml_opportunity_predictions
ORDER BY predicted_opportunity_score DESC
LIMIT 20;
```

```sql
-- Step 13: ML Model Performance Comparison

SELECT
  model_name,
  ROUND(rmse, 6) AS rmse,
  ROUND(mae, 6) AS mae,
  ROUND(r2, 6) AS r2,
  CASE
    WHEN model_name = 'Linear Regression' THEN 'Best Model'
    ELSE 'Comparison Model'
  END AS model_status
FROM workspace.census360.gold_ml_model_comparison
ORDER BY rmse ASC;
```

```sql
-- Step 14: Strategic Recommendation View

SELECT
  province_name,
  district_name,
  ds_division_name,
  gn_division_name,
  population_total,
  occupied_housing_units,
  ROUND(mobile_data_demand_score, 2) AS mobile_data_demand_score,
  ROUND(home_broadband_opportunity_score, 2) AS home_broadband_opportunity_score,
  ROUND(youth_package_opportunity_score, 2) AS youth_package_opportunity_score,
  ROUND(rural_coverage_priority_score, 2) AS rural_coverage_priority_score,
  ROUND(telco_demand_opportunity_score, 2) AS telco_demand_opportunity_score,
  opportunity_tier,
  CASE
    WHEN mobile_data_demand_score >= home_broadband_opportunity_score
      AND mobile_data_demand_score >= youth_package_opportunity_score
      AND mobile_data_demand_score >= rural_coverage_priority_score
      THEN 'Prioritize mobile data packages and network capacity'

    WHEN home_broadband_opportunity_score >= mobile_data_demand_score
      AND home_broadband_opportunity_score >= youth_package_opportunity_score
      AND home_broadband_opportunity_score >= rural_coverage_priority_score
      THEN 'Prioritize home broadband and fixed wireless broadband'

    WHEN youth_package_opportunity_score >= mobile_data_demand_score
      AND youth_package_opportunity_score >= home_broadband_opportunity_score
      AND youth_package_opportunity_score >= rural_coverage_priority_score
      THEN 'Prioritize youth, student, and social media packages'

    WHEN rural_coverage_priority_score >= mobile_data_demand_score
      AND rural_coverage_priority_score >= home_broadband_opportunity_score
      AND rural_coverage_priority_score >= youth_package_opportunity_score
      THEN 'Prioritize rural coverage expansion and connectivity investment'

    ELSE 'General telco opportunity area'
  END AS strategic_recommendation
FROM workspace.census360.gold_gn_telco_scored_v2
WHERE opportunity_tier IN ('Very High', 'High')
ORDER BY telco_demand_opportunity_score DESC;
```
