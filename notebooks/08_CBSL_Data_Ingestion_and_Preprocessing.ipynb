```sql
%python
import pandas as pd

cbsl_excel_path = (
    "/Volumes/workspace/census360/cbsl_raw_files/"
    "Economic and Social Statistics - 2025_chapter_2_e.xlsx"
)

workbook = pd.ExcelFile(cbsl_excel_path, engine="openpyxl")

print(f"Total sheets: {len(workbook.sheet_names)}")

for sheet_number, sheet_name in enumerate(workbook.sheet_names, start=1):
    print(f"{sheet_number}: {sheet_name}")
```
